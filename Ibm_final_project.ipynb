```python


import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    # Join the tokens back into a string
    text = " ".join(tokens)
    return text


# Apply the preprocessing function to the 'title' and 'description' columns
df_ibm['title'] = df_ibm['title'].apply(preprocess_text)
df_ibm['description'] = df_ibm['description'].apply(preprocess_text)
```

```python
df_ibm.head()["title"]
```

```python
import pandas as pd
# Sentiment Analysis
from textblob import TextBlob

def get_sentiment(text):
  analysis = TextBlob(text)
  return analysis.sentiment.polarity  # Returns a value between -1 (negative) and 1 (positive)

df_ibm['sentiment_title'] = df_ibm['title'].apply(get_sentiment)
df_ibm['sentiment_description'] = df_ibm['description'].apply(get_sentiment)
```

```python
df_ibm.head()
```

```python
# Analyze sentiment trends over time (if you have a date column)
# Assuming you have a 'date' column in your DataFrame
# Convert the date column to datetime objects if it's not already
df_ibm['date'] = pd.to_datetime(df_ibm['publish_date'])

#Group by month and calculate the average sentiment
daily_sentiment = df_ibm.groupby(pd.Grouper(key='date', freq='D'))[['sentiment_title', 'sentiment_description']].mean()
```

```python

# Combine 'title' and 'description' columns
df_ibm['title_des'] = df_ibm['title'] + ' ' + df_ibm['description']
df_ibm.head()
```

```python
df_ibm.to_csv("preprocessed_ibm.csv", index=False)
```

```python


# Topic Modeling (using Latent Dirichlet Allocation - LDA)
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Create a document-term matrix
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = vectorizer.fit_transform(df_ibm['title_des'])  # Use the description or title

# Apply LDA
LDA = LatentDirichletAllocation(n_components=5, random_state=42)  # Adjust n_components as needed
LDA.fit(dtm)

top_words = {}

# Get the top words for each topic
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}')
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
    print('\n')
    top_words[index] = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-15:]]


top_words = pd.DataFrame(top_words)
top_words.to_csv("top_words_ibm.csv", index=False)

```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from textblob import TextBlob
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# df_ibm = pd.read_csv("ibm_news.csv")
df_sector = pd.read_csv("it_sector_news.csv")


nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

def preprocess_text(text):
    # Check if the input is a string or can be converted to a string
    if isinstance(text, str) or not pd.isna(text):
        # Convert to string if not already
        text = str(text)
        # Lowercase the text
        text = text.lower()
        # Remove numbers
        text = re.sub(r'\d+', '', text)
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        # Tokenize the text
        tokens = word_tokenize(text)
        # Remove stop words
        stop_words = set(stopwords.words('english'))
        tokens = [w for w in tokens if not w in stop_words]
        # Join the tokens back into a string
        text = " ".join(tokens)
    # If it's a NaN value, replace with an empty string
    else:
        text = ''

    return text

# Apply the preprocessing function to the 'title' and 'description' columns
df_sector['title'] = df_sector['title'].apply(preprocess_text)
df_sector['description'] = df_sector['description'].apply(preprocess_text)
```
